## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
#@title 1.8 Find the norm of g and the norm of h and longest.
# Find the norm of two vectors and determine which is longer

g = np.array([1, 1, 1, 8])
h = np.array([3, 3, 3, 3])

g_norm = np.linalg.norm(g)
h_norm = np.linalg.norm(h)
if g_norm > h_norm:
    txt = 'g'
elif h_norm > g_norm:
    txt = 'h'

print('The norm of g is ', g_norm)
print('The norm of h is ', h_norm)
print('Therefore', txt, 'is longest')
```
